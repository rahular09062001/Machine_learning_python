```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
from sklearn import linear_model
%matplotlib inline

df=pd.read_csv('FuelConsumption.csv')
cdf=df[['FUELCONSUMPTION_COMB','CO2EMISSIONS','ENGINESIZE']]

msk=np.random.rand(len(df))<0.8
train=cdf[msk]
test=cdf[~msk]

train_x=np.asanyarray(train[['FUELCONSUMPTION_COMB','ENGINESIZE']])
train_y=np.asanyarray(train[['CO2EMISSIONS']])

regr=linear_model.LinearRegression()
regr.fit(train_x,train_y)

print("Coefficent : ",regr.coef_)
print("Intercept  : ",regr.intercept_)

test_x=np.asanyarray(test[['FUELCONSUMPTION_COMB','ENGINESIZE']])
test_y=np.asanyarray(test[['CO2EMISSIONS']])
test_y_=regr.predict(test_x)

plt.scatter(test_x[:,1],test_y_,color='green')
plt.scatter(test_x[:,1],test_y,color='orange')
```
